# Third Phase (Recognition Character)

```python
character = ['ا','ب','ت','ث','ج','ح','خ','د','ذ','ر','ز','س','ش','ص','ض','ط','ظ','ع','غ','ف','ق','ك','ل','م','ن','هـ','و','ي']
```

```python
import numpy as np
# upload the model
from tensorflow.keras import models
# this saved model
model = models.load_model("/content/drive/MyDrive/Full_Model/best_model.h5")

import pandas as pd
# upload testing data
test = pd.read_csv("/content/csvTestImages 3360x1024.csv")
# upload testing labels
test_label = pd.read_csv("/content/csvTestLabel 3360x1.csv")
# convert to array by numby 
import numpy as np
Test = np.array(test)
Test_label = np.array(test_label)
# i don't need to do reshape becouse the inpute is 1D not 2D
# convert the label to vector from 0 and the result cel is 1
from sklearn.preprocessing import LabelBinarizer
binencoder = LabelBinarizer()
Test_label = binencoder.fit_transform(Test_label)
# evaluate the model.
scores = model.evaluate(Test.reshape(-1,32,32), Test_label, verbose=0)
# to print the result 
scores
```

```python
#predict the result
result = model.predict(Test[200].reshape(-1,32,32))
```

```python
character[np.argmax(result)]
```

# Third Phase (Recognition Numbers)

```python
import numpy as np
# upload the model
from tensorflow.keras import models
# this saved model
model = models.load_model("/content/drive/MyDrive/Full_Model/Number Model.h5")

import pandas as pd
# upload testing data
test = pd.read_csv("/content/y.csv")
# upload testing labels
test_label = pd.read_csv("/content/y-label.csv")
# convert to array by numby 
import numpy as np
Test = np.array(test)
Test_label = np.array(test_label)
# i don't need to do reshape becouse the inpute is 1D not 2D
# convert the label to vector from 0 and the result cel is 1
from sklearn.preprocessing import LabelBinarizer
binencoder = LabelBinarizer()
Test_label = binencoder.fit_transform(Test_label)
# evaluate the model.
scores = model.evaluate(Test, Test_label, verbose=0)
# to print the result 
scores
```

```python
result = model.predict(Test[20].reshape(-1,784))
```
